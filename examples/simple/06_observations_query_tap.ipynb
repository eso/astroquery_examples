# **Observational Data: Query using TAP**

The Table Access Protocol (TAP) allows you to perform advanced searches on ESO's Science Archive using the Astronomical Data Query Language (ADQL).

You can query different tables depending on your science goal:

- **`TAP_SCHEMA.tables`** — metadata about all available tables and their columns
- **`ivoa.ObsCore`** — standardized metadata for fully calibrated Phase 3 science-ready products
- **`dbo.raw`** — metadata for raw observational data (spanning all ESO instruments)
- **`ist.<instrument_name>`** — instrument-specific metadata tables (e.g., `ist.midi`, `ist.muse`), offering fine-grained control
- **`asm.lhatpros_paranal_profiles`** - atmospheric profiles from the LHATPRO microwave radiometer at Paranal 

Additional examples are available in the ESO documentation:

- https://archive.eso.org/tap_obs/examples  
- https://archive.eso.org/programmatic/#TAP

By default, `eso.query_tap()` connects to the observations TAP service (`tap_obs`). 

Note that this can also be provided explicitly via the `tap_endpoint` argument:
```python
eso.query_tap("...", tap_endpoint="tap_obs") 
```

# **Query the ESO Archive using the TAP Service**

As a first step, it is useful to know what tables are available to query. This can be done using the following ADQL query, which queries the `TAP_SCHEMA.tables` metadata table:

````python
adql_query = """
            SELECT table_name
            FROM TAP_SCHEMA.tables
            ORDER BY table_name
            """

eso.query_tap(adql_query)
````

## **Reduced data search (ivoa.ObsCore)**

This example shows how to query the **VEGAS** survey (VST Early-type GAlaxieS Survey) for **science-ready, reduced imaging** of bright early-type galaxies from the NGC catalog. These data products are part of ESO's **Phase 3 collection** and include calibrated FITS images.

We filter the query for:
- `data_collection = 'VEGAS'`
- Target names from the NGC catalog
- Spatial resolution constraints

Such products are ideal for studying galaxy morphology and low-surface-brightness features.

First it is always useful to check which columns are available in the table you want to query. This can be done using ``eso.query_surveys(help=True)``, which queries `ivoa.ObsCore`.

````python
eso.query_surveys(help=True)
````

Which can also be done using the following ADQL query, which queries the `TAP_SCHEMA.columns` metadata table for columns in `ivoa.ObsCore`:


````python
adql_query = """
            SELECT column_name, datatype, xtype, unit, ucd, utype
            from TAP_SCHEMA.columns
            where table_name='ivoa.ObsCore'
            """

eso.query_tap(adql_query)
````

Now we can perform the actual query for VEGAS data in `ivoa.ObsCore`:

````python
adql_query = """
            SELECT TOP 30 target_name, s_ra, s_dec, instrument_name, 
                obs_collection, dataproduct_type, s_pixel_scale, 
                calib_level, t_min, proposal_id, dp_id
            FROM ivoa.ObsCore
            WHERE obs_collection = 'vegas'
                AND calib_level = 3
                AND dataproduct_type = 'image'
                AND s_pixel_scale < 0.4
                AND target_name LIKE 'NGC%'
            ORDER BY t_min DESC
            """

table = eso.query_tap(adql_query)
table["target_name", "s_ra", "s_dec", "instrument_name", "dataproduct_type", "s_pixel_scale", "dp_id"][:5]
````

## **Raw data search (dbo.raw)**

This query targets **raw narrow-band imaging** around Hα (656.3 nm) for the galaxy **NGC 3627**.

Key filters applied:
- Cone search around galaxy coordinates (RA/Dec) with 0.1° radius
- Spectral band includes Hα
- Filter width < 20 nm to exclude broadband observations
- Only science exposures (`dp_cat = 'SCIENCE'`)

These raw frames are useful for custom data reduction workflows focused on ionized gas and star formation in nearby galaxies.

````python
from astropy.coordinates import SkyCoord # import the SkyCoord class from the astropy.coordinates module

target = "NGC 3627" # set the target to NGC 3627
coords = SkyCoord.from_name(target) # create a SkyCoord object from the name of the source
````

Again, before running the query, it is useful to check which columns are available in the table you want to query. This can be done using ``eso.query_main(help=True)``, which queries `dbo.raw`.

````python
eso.query_main(help=True)
````

Which can also be done using the following ADQL query, which queries the `TAP_SCHEMA.columns` metadata table for columns in `dbo.raw`:


````python
adql_query = """
            SELECT column_name, datatype, xtype, unit, ucd, utype
            from TAP_SCHEMA.columns
            where table_name='dbo.raw'
            """

eso.query_tap(adql_query)
````

Now we can perform the actual query for VEGAS data in `dbo.raw`:

````python
adql_query = f"""
            SELECT *
            FROM dbo.raw
            WHERE dp_cat = 'SCIENCE'
                AND 1 = INTERSECTS(s_region, CIRCLE('', {coords.ra.value},  {coords.dec.value}, 0.1))  -- defining a cone of 0.1 deg radius around NGC 3627
                AND lambda_min < 656 
                AND lambda_max > 656                                -- Ensure Hα falls in the filter bandpass
                AND lambda_max - lambda_min < 20                    -- Narrow-band filter (~< 20 nm)
            ORDER BY mjd_obs
            """ 

table = eso.query_tap(adql_query)
table["object", "ra", "dec", "instrument", "exposure", "tel_ambi_fwhm_start", "date_obs", "prog_id", "dp_id", "access_url"][:5]
````

As with ``eso.query_main`` vs ``eso.query_instrument`` we can also query the instrument specific tables directly in tap using the `ist.[instrument]` syntax.

For example in this case we can, for example, query the `ist.muse` table directly to get observations from **MUSE** (IFU) only, but with more instrument specific columns available.

As always, its good to check which columns are available in the table you want to query. This can be done using ``eso.query_instrument("MUSE", help=True)``, which queries `ist.muse`.

# **Cone search using TAP**

Suppose you are looking for reducted datasets closer than 2.5 arcmin from NGC 4666.

Either you know the equatorial coordinates of your object, or you rely onto a name resolver like SESAME (CDS) to get them. 

In the end you have a circle defined by the 3 quantities; `ra`, `dec`, and `radius`, all expressed in degrees.

The cone search constraint is defined by the `INTERSECTS` operator, which takes two footprints as input, and return true (1) if the two intersects in at least 1 point, false (0) if they don't.

In a cone search query, one of the two footprints is the input circle, while the other is the `_s_region_` column, which represents the footprint of any dataset.

````python
from astropy import units as u # import the units module from astropy

# Defining the position via SESAME name resolver, and the search radius
target = "NGC 4666"

pos = SkyCoord.from_name(target)
# pos now contains the coordinates of NGC 4666
print("SESAME coordinates for %s: %s (truncated to millidegrees)\n" % (target, pos.to_string()))

radius = (2.5 *u.arcmin).to(u.degree) # search radius of 2.5 arcmin, always expressed in degrees

# Cone search: looking for footprints of reduced datasets intersecting a circle of 2.5' around NGC 4666
adql_query = """SELECT *
            FROM ivoa.ObsCore
            WHERE intersects(s_region, circle('', %f, %f, %f))=1
            """ % (pos.ra.degree, pos.dec.degree, radius.value)

table = eso.query_tap(adql_query)
table[:3]
````

## **Cone search: INTERSECTS or CONTAINS?**

If you are looking specifically for spectra (i.e. point-like footprints), you could use the stricter `CONTAINS` operator instead. This ensures that only footprints fully contained within the defined circle are returned. Note:
- `INTERSECTS` is commutative (order does not matter)
- `CONTAINS` is not commutative, and must be used as:
    `CONTAINS(contained, container) = 1`

You might want to repeat the cone search query after replacing `INTERSECTS` with `CONTAINS` to see the difference. Expected result: images (and measurements derived from them) no longer appear in the result table, as their larger field of view (> 2.5') means they are not fully contained within the search region.

# **Spatial joins using TAP**

Interested in finding images in different bands covering the same sky region for photometric studies?

The example below demonstrates how to construct a spatial join to identify:

- HAWK-I images  
- within 10 degrees of the Galactic plane  
- taken in the J and H filters  
- where the J and H images overlap  
- with at least 80% of the J-band image area covered by the overlap

````python
adql_query = """
            SELECT J.* FROM
            (select * FROM ivoa.Obscore WHERE dataproduct_subtype ='srctbl' 
                AND obs_collection = 'HAWKI' 
                AND gal_lat < 10 AND gal_lat > -10 
                AND em_min < 1.265E-6 AND em_max > 1.265E-6 ) J, 
                (select * FROM ivoa.Obscore WHERE dataproduct_subtype ='srctbl' 
                AND obs_collection = 'HAWKI' 
                AND gal_lat < 10 AND gal_lat > -10 
                AND em_min < 1.66E-6 AND em_max > 1.66E-6 ) H 
            WHERE INTERSECTS( J.s_region , H.s_region)=1 and 
            ESO_INTERSECTION( J.s_region , H.s_region) > 0.8*AREA( J.s_region )
            """

# Let's get a maximum of 20000 images
eso.ROW_LIMIT = 20000
table = eso.query_tap(query=adql_query)
eso.ROW_LIMIT = None

print("")
print("Num matching datasets: %d" % (len(table)))
````

Notice the `J.*` in the `SELECT` part: that is to retrieve just only the information about the J images. You could then repeat the query for the H band. This is useful, for example, to visualise the results in a tool like Aladin, using different colours for the different queries, and hence for the different bands.

Otherwise, you could simply get all of them at once, using `SELECT *` instead, or using a more selective statement, prepending the column name that you want from the J or H set of results, like in: `SELECT J.dp_id, H.dp_id, etc`.

# **Query by wavelengths**

Note that the IVOA ObsCore standard requires wavelengths to be stored in meters.

You can convert to other units in the `SELECT` statement (as shown below), but any constraints on em_min and em_max must be specified in meters. This applies even to data with a frequency-based spectral axis (e.g. radio observations). Using meters ensures a consistent query interface across all observatories, so remember to apply constraints in meters.

````python
adql_query = """
            SELECT obs_collection as collection, dataproduct_type as type, 
            dataproduct_subtype as subtype, em_min*1E9 min_wavel_nm, em_max*1E9 max_wavel_nm, em_res_power
            FROM ivoa.ObsCore
            WHERE target_name = 'a370'
            and em_res_power < 3000
            """

eso.ROW_LIMIT = 3
table = eso.query_tap(query=adql_query)
eso.ROW_LIMIT = None
table
````

# **Atmospheric profiles (`asm.lhatpros_paranal_profiles`)**

One advantage of using the TAP query interface directly is that it provides access to tables that are not (currently) exposed through the higher-level `astroquery.eso` convenience functions. One such example is the `asm.lhatpros_paranal_profiles` table, which contains atmospheric profiles measured at Cerro Paranal by the LHATPRO microwave radiometers. Profiles are available at user-specified sky coordinates and at 39 altitude levels from 0 to 10 km above the VLT platform, and include absolute humidity, relative humidity, and temperature (typically averaged over ~60 s; archive times are in UTC).

In the cells below, we query profiles within a cone of radius `radius` (degrees) centred on (`ra`, `dec`) and within a midpoint time window of ±`within` seconds around `midpoint_iso`, then plot the vertical profiles versus altitude, with individual time samples colour-coded and the median profile highlighted.

````python
# Cone search parameters (degrees)
ra_deg = 332.2
dec_deg = -25.5
radius_deg = 2.5

# Time window: midpoint ± 45 minutes
midpoint_iso = "2019-06-06T09:55:27"
window_minutes = 45
window_seconds = window_minutes * 60

adql_query = f"""
            SELECT start_date, midpoint_date, ra, dec, alt, az, *
            FROM asm.lhatpros_paranal_profiles
            WHERE midpoint_date BETWEEN eso_dateadd_sec(-{window_seconds}, '{midpoint_iso}')
                                AND eso_dateadd_sec( {window_seconds}, '{midpoint_iso}')
            AND CONTAINS(
                    POINT('', ra, dec),
                    CIRCLE('', {ra_deg}, {dec_deg}, {radius_deg})
                ) = 1
            AND valid = 1
            """

table = eso.query_tap(adql_query)
table[:5]
````

````python
import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time

# Plot absolute humidity, relative humidity, and temperature profiles across altitude
if len(table) == 0:
    print("No rows returned for the time/position window.")
else:
    altitudes_m = np.array([
        0, 10, 30, 50, 75, 100, 125, 150, 200, 250, 325, 400, 475, 550,
        625, 700, 800, 900, 1000, 1150, 1300, 1450, 1600, 1800, 2000,
        2200, 2500, 2800, 3100, 3500, 3900, 4400, 5000, 5600, 6200,
        7000, 8000, 9000, 10000
    ], dtype=float)
    N_LEVELS = len(altitudes_m)

    def build_profiles(prefix):
        """Extract profile columns like f'{prefix}1'.. and return (alt_m, profiles).

        Returns
        -------
        alt_m : ndarray, shape (n_levels_found,)
        profiles : ndarray, shape (n_times, n_levels_found)
        """
        cols = [f"{prefix}{i}" for i in range(1, N_LEVELS + 1)]
        cols = [c for c in cols if c in table.colnames]

        if not cols:
            return altitudes_m[:0], np.empty((len(table), 0), dtype=float)

        alt_m = altitudes_m[:len(cols)]
        data = []
        for col in cols:
            values = np.array(table[col])
            if hasattr(values, "filled"):  # masked arrays
                values = values.filled(np.nan)
            data.append(values)

        profiles = np.array(data, dtype=float).T
        return alt_m, profiles

    alt_ahum, ahum = build_profiles("ahum")
    alt_rhum, rhum = build_profiles("rhum")
    alt_temp, temp = build_profiles("temp")

    # Sort rows by midpoint time
    times = Time(list(table["midpoint_date"].data.data))
    order = np.argsort(times.unix)
    times_dt = np.array(times.to_datetime())[order]
    ahum, rhum, temp = ahum[order], rhum[order], temp[order]

    fig, axes = plt.subplots(1, 3, figsize=(12, 5), sharey=True, constrained_layout=True)
    cmap = plt.cm.viridis
    norm = plt.Normalize(0, max(len(times_dt) - 1, 1))

    def draw_panel(ax, profiles, altitudes, xlabel, title):
        # profiles has shape (n_times, n_levels)
        for i, profile in enumerate(profiles):
            ax.plot(profile, altitudes, color=cmap(norm(i)), alpha=0.4, lw=1)
        if profiles.size:
            median_profile = np.nanmedian(profiles, axis=0)
            ax.plot(median_profile, altitudes, color="black", lw=3, label="Median")
        ax.set_xlabel(xlabel)
        ax.set_title(title)
        ax.grid(True, alpha=0.2, ls=":")

    draw_panel(axes[0], ahum, alt_ahum, r"Absolute humidity [g/m$^3$]", "Absolute humidity")
    draw_panel(axes[1], rhum, alt_rhum, "Relative humidity [%]", "Relative humidity")
    draw_panel(axes[2], temp-273.15, alt_temp, "Temperature [°C]", "Temperature")

    axes[0].set_ylabel("Altitude [m]")
    axes[2].legend(frameon=False, loc="lower left")

    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=axes, pad=0.02, shrink=0.85)

    if len(times_dt) > 1:
        cbar.set_ticks([0, len(times_dt) - 1])
        cbar.set_ticklabels([times_dt[0].strftime("%H:%M:%S"), times_dt[-1].strftime("%H:%M:%S")])
        cbar.set_label("Sorted midpoint time (earliest → latest)")
    else:
        cbar.set_label("Midpoint time order")
````
